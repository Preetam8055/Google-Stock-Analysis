```python
   import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Download stock data
stock = 'GOOG'  # Google stock ticker symbol
start = '2020-01-01'  # Start date
end = '2024-01-01'  # End date
google_data = yf.download(stock, start=start, end=end)

# Function to plot various graphs
def plot_graphs(figsize, data, column_name):
    fig, axes = plt.subplots(2, 2, figsize=figsize)

    # Bar Graph
    axes[0, 0].bar(data.index, data)
    axes[0, 0].set_title(f'Bar Graph of {column_name}')
    axes[0, 0].set_xlabel('Date')
    axes[0, 0].set_ylabel(column_name)
    
    # Histogram
    axes[0, 1].hist(data, bins=30, color='blue', alpha=0.7)
    axes[0, 1].set_title(f'Histogram of {column_name}')
    axes[0, 1].set_xlabel(column_name)
    axes[0, 1].set_ylabel('Frequency')

    # Box Plot
    sns.boxplot(ax=axes[1, 0], data=data)
    axes[1, 0].set_title(f'Box Plot of {column_name}')
    
    # Scatter Plot
    axes[1, 1].scatter(np.arange(len(data)), data, c='green', alpha=0.6)
    axes[1, 1].set_title(f'Scatter Plot of {column_name}')
    axes[1, 1].set_xlabel('Index')
    axes[1, 1].set_ylabel(column_name)

    # Rotate x-axis labels for bar graph
    for tick in axes[0, 0].get_xticklabels():
        tick.set_rotation(45)

    plt.tight_layout()
    plt.show()

# Function for pie chart
def plot_pie_chart(data, column_name):
    plt.figure(figsize=(6, 6))
    data_values = data[-5:]  # Get the last 5 values for pie chart
    data_values.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart of Last 5 {column_name} values')
    plt.ylabel('')
    plt.show()

# Plot each graph for each column
for column in google_data.columns:
    plot_graphs((15, 10), google_data[column], column)
    plot_pie_chart(google_data[column], column)
```
